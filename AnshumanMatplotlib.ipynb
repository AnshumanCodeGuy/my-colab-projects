```python
import matplotlib.pyplot as plt
name=["Anshuman","Sumit","Ravi","Amit","Pappu"]
marks=[88,89,90,85,84]
plt.bar(name,marks,color="red",width=.4)
plt.xlabel("Name",fontsize=15)
plt.ylabel("Marks",fontsize=15)
plt.title("Student Marks",fontsize=20)
plt.grid()
plt.show()
```
